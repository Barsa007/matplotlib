```python
import numpy as np 
import pandas as pd
from IPython.display import display

import matplotlib.pyplot as plt
```

```python
'''
Create x = np.arange(0, 11) to represent 0‚Äì10 kilometers
Compute y = 8 + 5 * x for the taxi fare
Plot the line with: title, x/y labels, and legend
Limit the axes to xlim(0, 10) and ylim(0, 60)
Save as taxi_fare_line.jpg and show the plot
'''
#1
x = np.arange(0, 11)

#2
y = 8 + 5 * x

#3
fig, graph= plt.subplots(nrows=1,ncols=1)
graph.plot(x, y,label='Taxi Fare vs Distance',color='green')
graph.set_xlabel('Kilometers')
graph.set_ylabel('Taxi Fare')
graph.legend()

#4
graph.set_xlim(0, 10)
graph.set_ylim(0, 60)
fig.suptitle('1.Taxi Fare vs Distance',x=0.5)

#5
plt.savefig('taxi_fare_line.jpg')
plt.show()
```

```python
display('2. Visual Analysis with tips.csv üçΩÔ∏è')
df = pd.read_csv("tips.csv")
'''
2.1 Scatter: price_per_person (x) vs tip (y)
Create a scatter plot of price_per_person (x-axis) vs tip (y-axis)
Add title and axis labels
By looking at the graph- Is there a correlation?
'''


x=df.sort_values('price_per_person',ascending=True)['price_per_person']
y=df.sort_values('price_per_person',ascending=True)['tip']
plt.scatter(x, y, color='blue')
plt.title('2.1 price_per_person vs tip')
plt.xlabel('price_per_person')
plt.ylabel('tip')
plt.legend(['8==D'])
plt.show()
```

```python
'''
2.2. Bar: maximum total_bill per day
For each day, find the max total_bill
Plot a bar chart of these maxima
'''
display('2.2. Bar: maximum total_bill per day')
max_bill_per_day=df.groupby('day')['total_bill'].max()
plt.bar(max_bill_per_day.index, max_bill_per_day.values)
```

```python
'''2.3. Histogram: tips
Plot a histogram of tip
Question:By looking at the graph- what is the range of the most common tip?'''
'''
0!~ -0.5
'''
'''
2.4. Histogram: tip percentage
Create a new column tip_perc = 100 * tip / total_bill
Plot a histogram of tip_perc
Question: By looking at the graph- what is the range of the most tip percentage?
'''
billdf=df.sort_values('total_bill',ascending=True)
tip_perc = 100 * billdf['tip'] / billdf['total_bill']
plt.hist(tip_perc, bins=10, color='green')
plt.show()
#10~20
```

```python
'''2.5. Subplots: combine all four visuals
Create a 2√ó2 grid using fig, axes = plt.subplots(2, 2, figsize=(10, 7))
Recreate the four plots in their own axes objects: (1) scatter, (2) bar of max per day, (3) tip histogram, (4) tip% histogram
Make sure each subplot has a title and labeled axes
Adjust layout and save as tips_overview.png'''

axes ,graph= plt.subplots(2, 2, figsize=(10, 7))

#scatter
x0=df.sort_values('price_per_person',ascending=True)['price_per_person']
y0=df.sort_values('price_per_person',ascending=True)['tip']
axes.suptitle('2.5. Subplots',x=0.5)
graph[0,0].set_title("Scatter")
graph[0,0].set_ylabel('Tip')
graph[0,0].scatter(x0,y0,label ='Price per person Vs Tip',color='blue')
graph[0,0].legend()

#bar of max per day
graph[0,1].set_title("bar of max per day")
graph[0,1].set_ylabel('Max Tottal Bill')
graph[0,1].bar(max_bill_per_day.index, max_bill_per_day.values,color='orange')


#tip histogram
tip=df['tip']
graph[1,0].set_title("tip histogram")
graph[1,0].set_ylabel('tip amount')
graph[1,0].hist(tip, bins=10, color='red')

#tip% histogram
billdf=df.sort_values('total_bill',ascending=True)
tip_perc = 100 * billdf['tip'] / billdf['total_bill']
graph[1,1].set_title("tip% histogram")
graph[1,1].set_ylabel('tip% amount')
graph[1,1].hist(tip_perc, bins=10, color='green')
```
