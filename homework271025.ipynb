```python
#bars homework

# For each of those linear systems, solve and find if the systems 
# have single solution, no solution or infinitely solutions.
#I.    x+y=12 
#      x-y=4
#
#II.  5x + 3y = 13 
#     4x - 2y = 6
#
#III. 2x + 4y = 8 
#     x + 2y = 4
#
```

```python
#solution #1
x= np.arange(-10,10)
y1=12-x 
y2=x-4

plt.title('Solution I - single solution')
plt.plot(x,y1,color='red',label="y1=12-x ")
plt.plot(x,y2,color='purple',label="y2=x-4")
plt.axhline(0,color='black',linewidth=0.8,linestyle='--')
plt.axvline(0,color='black',linewidth=0.8,linestyle='--')
plt.grid(color='grey',linewidth=0.5,linestyle='--')
#find courdinants
#12-x=x-4/+4+x
#2x=16
#x=8  
#(8,Y?)

#y=12-8
#y=4
(8,4)
plt.scatter(8,4,color='black',zorder=5,label='(8,4)')
plt.legend()

```

```python
x=np.arange(-10,10)
y1=4.333-1.6666*x
y2=2*x-3
plt.title('Solution II - single solution')
plt.plot(x,y1,color='green',label='y1=4.333-1.6666x')
plt.plot(x,y2,color='orange',label='y2=2x-3')
plt.axhline(0,color='black',linewidth=0.8,linestyle='--')
plt.axvline(0,color='black',linewidth=0.8,linestyle='--')
plt.grid(color='grey',linewidth=0.5,linestyle='--')

#find courdinants
#4.333-1.6666x=2x-3/+3/+1.6666x
#3.6666x=7.3333
#7.3333/3.6666=2.0000272732231497(roghly= 2)
#x=2
#y2=2*2-3
#y=1
plt.scatter(2,1,color='black',zorder=5,label='(2,1)')
plt.legend()
```

```python
x=np.arange(-10,10)
#2x + 4y = 8 /-2x
#4y=8-2x/:4
y1=2-0.5*x
#x + 2y = 4/-x/:2
y2=2-0.5*x #infinitely solutions.
plt.title('Solution III - infinitely solutions')
plt.plot(x,y1,color='yellow',label='y1=2-0.5x')
plt.plot(x,y2,color='blue',label='y2=2-0.5x')
plt.axhline(0,color='black',linewidth=0.8,linestyle='--')
plt.axvline(0,color='black',linewidth=0.8,linestyle='--')
plt.grid(color='grey',linewidth=0.5,linestyle='--')
plt.legend()
```
